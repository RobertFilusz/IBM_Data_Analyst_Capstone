## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(y = df["ConvertedComp"])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
ql = df['ConvertedComp'].quantile(0.25)
qu = df['ConvertedComp'].quantile(0.75)
iqr = qu - ql
print('inter quartile range:', iqr)
```

Find out the upper and lower bounds.


```python
# your code goes here
lower = df['ConvertedComp'].min()
upper = df['ConvertedComp'].max()
print('upper:', upper, 'lower', lower)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
print('Outliers below:',df['ConvertedComp'].lt(ql - 1.5*iqr).sum())
print('Outliers above:',df['ConvertedComp'].gt(qu + 1.5*iqr).sum())
print('Median after removing outliers:',df[df['ConvertedComp'].le(qu + 1.5*iqr)]['ConvertedComp'].median())
print('Mean after removing outliers:',df[df['ConvertedComp'].le(qu + 1.5*iqr)]['ConvertedComp'].mean())
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_new = df[df['ConvertedComp'].le(qu + 1.5*iqr)]
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.
